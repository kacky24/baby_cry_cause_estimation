```python
%matplotlib inline
import soundfile as sf
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
```

データ読み込み

```python
data, sample_rate = sf.read('sample2.wav')
x = [t / 44100 for t in range(data.shape[0])]
plt.plot(x, data[:, 0])
plt.show()
```

```python
sf.info('sample3.wav')
```

welchメソッドによるパワースペクトル獲得

```python
import scipy.signal
fs = 12000
nperseg = 128
freq, P = scipy.signal.welch(data[:,0], fs, window='hamming', nperseg=nperseg, nfft=1024)
```

描写

```python
plt.plot(freq, 10*np.log10(P))
plt.xlabel("Frequency[Hz]")
plt.ylabel("Power/frequency[dB/Hz]")
plt.show()
```
