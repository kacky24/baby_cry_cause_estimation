```python
%matplotlib inline
import soundfile as sf
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
```

データ読み込み

```python
# data, sample_rate = sf.read('sample_data/Louise_01.m4a_0.wav')
data, sample_rate = sf.read('output.wav')
#x = [t / 44100 for t in range(data.shape[0])]
plt.plot(data)
plt.show()
```

```python
sf.info('Louise_01.m4a_0.wav')
```

welchメソッドによるパワースペクトル獲得

```python
import scipy.signal
fs = 22000
nperseg = 128
freq, P = scipy.signal.welch(data, fs, window='hamming', nperseg=nperseg, nfft=1024)
```

描写

```python
plt.plot(freq, 10*np.log10(P))
plt.xlabel("Frequency[Hz]")
plt.ylabel("Power/frequency[dB/Hz]")
plt.show()
```

```python
f, t, Sxx = scipy.signal.spectrogram(data,fs=44100, nperseg=1024)
```

```python
plt.figure()
plt.pcolormesh(t,f,Sxx,vmax=1e-6)
#plt.xlim([0,18])
plt.xlabel(u"time [sec]")
plt.ylabel(u"freqency [Hz]")
# plt.colorbar()
plt.show()
```
